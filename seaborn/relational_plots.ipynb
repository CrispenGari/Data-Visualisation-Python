### 2. Relational plots

### 2.1 relplot
* Figure-level interface for drawing relational plots onto a FacetGrid.
```
seaborn.relplot(*, x=None, y=None, hue=None, size=None, style=None, data=None, row=None, col=None, col_wrap=None, row_order=None, col_order=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=None, dashes=None, style_order=None, legend='auto', kind='scatter', height=5, aspect=1, facet_kws=None, units=None, **kwargs)
```
> Find kwargs in the [docs](https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot)

* Seaborn comes with some datasets in it which I will be using here

````python
tipsDf = sns.load_dataset("tips")
tipsDf.tail(10).reset_index()
````

````python
sns.relplot(data=tipsDf, x="total_bill", y="tip", hue="day", kind="scatter")
plt.show()
````

````python
# sns.set_theme()
sns.relplot(data=tipsDf, x="total_bill", y="tip", hue="smoker", kind="scatter", col="time")
plt.show()
````

````python
sns.relplot(data=tipsDf, x="total_bill", y="tip", hue="smoker", kind="scatter", row="sex", col="time")
plt.show()
````

````python
print(tipsDf.shape)
tipsDf.head()
````

````python
sns.relplot(data=tipsDf, x="total_bill",y="tip" ,hue="time", col="day", col_wrap=2)
plt.show()
````

````python
fmri = sns.load_dataset("fmri")
fmri.head()
````

````python
sns.relplot(data=fmri, x="timepoint", y="signal", hue="event", col="region", kind="line", style="subject")
plt.show()
````

````python
flights_wide = sns.load_dataset("flights").pivot("year", "month", "passengers")
````

````python
flights_wide.head()
````

````python
sns.relplot(data=flights_wide, kind="line")
plt.show()
````

````python
a = np.random.rand(100)
b = np.array([i**2 for i in a])

sns.relplot(x=a, y=b, hue=b)
plt.show()
````

### 2.2. scatterplot
```seaborn.scatterplot(*, x=None, y=None, hue=None, style=None, size=None, data=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=True, style_order=None, x_bins=None, y_bins=None, units=None, estimator=None, ci=95, n_boot=1000, alpha=None, x_jitter=None, y_jitter=None, legend='auto', ax=None, **kwargs)```
* Draw a scatter plot with possibility of several semantic groupings.

````python
x = np.random.rand(50)
y = np.random.rand(50)
````

````python
sns.scatterplot(x=x, y=y)
plt.show()
````

````python
tips = sns.load_dataset("tips")
tips.tail().reset_index()
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time")
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size")
plt.show()
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", palette="deep")
plt.show()
````

````python
tip_rate = tips.eval("tip / total_bill").rename("tip_rate")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue=tip_rate,)
plt.show()
````

````python

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time" , size="size", sizes=(20, 100), markers={"Lunch": 'o', "Dinner": 's'})
plt.show()
````

````python
markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(data=tips, x="total_bill", y="tip", style="time", markers=markers)
````

````python
index = pd.date_range("1 1 2000", periods=100, freq="m", name="date")
data = np.random.randn(100, 4).cumsum(axis=0)
wide_df = pd.DataFrame(data, index, ["a", "b", "c", "d"])
sns.scatterplot(data=wide_df)
````
