## Distribution plots

````python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
````

### 1. displot
* Figure-level interface for drawing distribution plots onto a FacetGrid.
```
seaborn.displot(data=None, *, x=None, y=None, hue=None, row=None, col=None, weights=None, kind='hist', rug=False, rug_kws=None, log_scale=None, legend=True, palette=None, hue_order=None, hue_norm=None, color=None, col_wrap=None, row_order=None, col_order=None, height=5, aspect=1, facet_kws=None, **kwargs)
```
* [Docs](https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot)

````python
penguins = sns.load_dataset("penguins")
penguins[:2]
````

````python
sns.displot(data=penguins, x="bill_length_mm", hue="sex")
plt.show()
````

````python
sns.displot(data=penguins, x="bill_length_mm", hue="sex", kind="kde")
plt.show()
````

````python
sns.displot(data=penguins, x="bill_length_mm", hue="sex", kind="ecdf")
plt.show()
````

````python
sns.displot(data=penguins, x="bill_length_mm", hue="sex", kde=True)
plt.show()
````

````python
sns.displot(data=penguins, x="bill_length_mm", hue="sex", y="flipper_length_mm")
plt.show()
````

````python
sns.displot(data=penguins, x="bill_length_mm", hue="sex", y="flipper_length_mm", kind="kde")
plt.show()
````

````python
sns.displot(data=penguins, hue="species", x="flipper_length_mm", kind="kde")
plt.show()
````

````python
sns.displot(data=penguins, x="bill_length_mm", hue="sex", col="species", kind="kde")
plt.show()
````

### 2. histplot()
* Plot univariate or bivariate histograms to show distributions of datasets.
```
eaborn.histplot(data=None, *, x=None, y=None, hue=None, weights=None, stat='count', bins='auto', binwidth=None, binrange=None, discrete=None, cumulative=False, common_bins=True, common_norm=True, multiple='layer', element='bars', fill=True, shrink=1, kde=False, kde_kws=None, line_kws=None, thresh=0, pthresh=None, pmax=None, cbar=False, cbar_ax=None, cbar_kws=None, palette=None, hue_order=None, hue_norm=None, color=None, log_scale=None, legend=True, ax=None, **kwargs)
```

````python
sns.histplot(data=penguins, x="flipper_length_mm", bins=10, kde=True)
plt.show()
````

````python
sns.histplot(data=penguins, y="flipper_length_mm", bins=10, kde=True)
plt.show()
````

````python
sns.histplot(data=penguins, x="flipper_length_mm", bins=10, kde=False, hue="species")
plt.show()
````

````python
sns.histplot(data=penguins, x="flipper_length_mm", bins=10, fill=False, element="step")
plt.show()
````

````python
sns.histplot(data=penguins, x="flipper_length_mm", bins=10, element="poly", hue="sex")
plt.show()
````

### 3. 	 kdeplot()
* Plot univariate or bivariate distributions using kernel density estimation.

```
seaborn.kdeplot(x=None, *, y=None, shade=None, vertical=False, kernel=None, bw=None, gridsize=200, cut=3, clip=None, legend=True, cumulative=False, shade_lowest=None, cbar=False, cbar_ax=None, cbar_kws=None, ax=None, weights=None, hue=None, palette=None, hue_order=None, hue_norm=None, multiple='layer', common_norm=True, common_grid=False, levels=10, thresh=0.05, bw_method='scott', bw_adjust=1, log_scale=None, color=None, fill=None, data=None, data2=None, **kwargs)
```

````python
iris = sns.load_dataset("iris")
iris[:2]
````

````python
sns.kdeplot(data=iris, x="sepal_length")
plt.show()
````

````python
sns.kdeplot(data=iris, y="sepal_length")
plt.show()
````

````python
sns.kdeplot(data=iris, x="sepal_length", hue="species")
plt.show()
````

````python
sns.kdeplot(data=iris, x="sepal_length", hue="species", bw_adjust=.2)
plt.show()
````

````python
sns.kdeplot(data=iris, x="sepal_length", hue="species", bw_adjust=.2, fill=True)
plt.show()
````

````python
sns.kdeplot(data=iris, x="sepal_length", hue="species", bw_adjust=.2, multiple="stack", fill=True)
plt.show()
````

### 4. ecdfplot

```
seaborn.ecdfplot(data=None, *, x=None, y=None, hue=None, weights=None, stat='proportion', complementary=False, palette=None, hue_order=None, hue_norm=None, log_scale=None, legend=True, ax=None, **kwargs)
```

````python
sns.ecdfplot(data=penguins, hue="species", x="flipper_length_mm")
plt.plot()
````

### 5. rugplot
* Plot marginal distributions by drawing ticks along the x and y axes.
```
seaborn.rugplot(x=None, *, height=0.025, axis=None, ax=None, data=None, y=None, hue=None, palette=None, hue_order=None, hue_norm=None, expand_margins=True, legend=True, a=None, **kwargs)¶
```

````python
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
plt.show()
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip")
plt.show()
````

````python
sns.rugplot(data=tips, x="total_bill", y="tip")
plt.show()
````
