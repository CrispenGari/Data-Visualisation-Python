To complete the task A file name `complete-cohort-life-tables-1876-2018.csv` is located in the `data` folder. It contains some data the you are required to visualize using `matplotlib.pyplot`. Read this file in a pandas dataframe and check the last `3` columns in the dataframe using the `tail` method.

````python
dataframe = pd.read_csv('data/complete-cohort-life-tables-1876-2018.csv')
dataframe.tail(3)
````

In the following code cell you are required to clean the dataframe by removing `na`/`null` values.

````python
dataframe.dropna(inplace=True)
````

In the following code cell i will show you a simple example of ploting `count_plot` from a pandas `Series`. 

````python
plt.figure(figsize=(4, 4))
dataframe.sex.value_counts().plot(kind="pie")
plt.show()
````

You are required to read and understand the code that is in the following code cell. Change the `df` with a variable name of your dataframe

````python
xs = np.arange(len(dataframe))[:20]
plt.figure(figsize=(5, 3))
columns = np.array(['dx', 'ex', 'llx', 'lx', 'mx', 'px', 'qx','sx'])
colors = np.array(['#86A789', '#B2C8BA', '#D2E3C8', '#EBF3E8', '#3A4D39', 'black', 'yellow', 'green'])

for color, column  in zip(colors, columns) :
    plt.plot(xs, dataframe[column].values[:20], linewidth=.8, color=color)
plt.title('Line Graphs', fontsize=12)
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.legend(labels=columns, fontsize=8)
plt.show()
````

In the following code cell you are required to plot the first `30` values of `mx` aganist `px` as a scatter plot. Then on the same plot you are required to plot line of bestfit.

````python
mx = dataframe.mx.values[:30]
px = dataframe.px.values[:30]
c, m = polyfit(mx, px, 1)
ys = ys = m * mx +  c
plt.figure(figsize=(5, 3))
plt.plot(mx, ys, '-k', linewidth=1)
plt.scatter(mx, px)
plt.title(
    "mx vrs px", 
    fontfamily="Arial", 
    fontweight="bold", 
    fontsize=10
)
plt.xlabel(
    "mx",
    fontsize=8
)
plt.ylabel("px", fontsize=8)
plt.grid()
plt.show()
````

In the following code cell you are required to plot the first `20` values of `qx` aganist `sx`. For the plot colors you are required to use your custom colors.

````python
xs = np.arange(20)
qx = dataframe.qx.values[:20]
sx = dataframe.sx.values[:20]

plt.figure(figsize=(5, 3))
plt.scatter(xs, qx,  color='#0C356A', marker='o', edgecolors='k')
plt.scatter(xs, sx, color='#4F6F52', marker='o', edgecolors='g')
plt.title('Scatter Plot', fontsize=12)
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.grid()
plt.show()
````
