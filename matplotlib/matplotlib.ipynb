1. bargraph a.k.a (vertical bar)

The `plt.bar()` method allows us to plot the bar graph.

> You can get the bar clours from [colorhunt](https://colorhunt.co/).

```python
x = np.array(['Java', 'Kotlin', 'TypeScript', 'C++', 'Python'])
colors = np.array(['#86A789', '#B2C8BA', '#D2E3C8', '#EBF3E8', '#3A4D39'])
y = np.array([10, 20, 50, 3, 70])

plt.figure(figsize=(5, 3))
plt.bar(x, y, label=x, color=colors, width=.3)
plt.title('Programing Languages')
plt.xlabel("Programing Language", fontsize=8)
plt.ylabel("% likes", fontsize=8)
plt.legend()
plt.show()
```

Using the above data we want to plot a `horizontal` bar. To do that we will use a method calle `barh`

```python
plt.figure(figsize=(5, 3))
plt.barh(x, y, label=x, color=colors, height=.3)
plt.title('Programing Languages')
plt.ylabel("Programing Language", fontsize=8)
plt.xlabel("% likes", fontsize=8)
plt.legend()
plt.show()
```

2. linegraph

In the following code cell we are going to plot the line graph. To plot a line graph we use the `plot` method.

```python
x = y = np.arange(11)
plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=.8, linestyle='--', color='#0C356A', marker='o', markeredgecolor='red')
plt.plot(x**2, y, linewidth=.8, linestyle=':', color='#706233', marker='o', markeredgecolor='red')
plt.title('Line Graph', fontsize=12)
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.show()
```

3. scatterplot

In this section we are going to plot a scatter plot using  the `scatter` method.

```python
a = np.random.rand(10) * 100
b = np.random.rand(10) * 100


c = np.random.rand(10) * 100
d = np.random.rand(10) * 100

plt.figure(figsize=(5, 3))
plt.scatter(a, b,  color='#0C356A', marker='o', edgecolors='red')
plt.scatter(c, d, color='#4F6F52', marker='o', edgecolors='k')
plt.title('Scatter Plot', fontsize=12)
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.show()
```
