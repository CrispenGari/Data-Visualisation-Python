In the following code cell we are going to import the packages that we are going to use in this notebook.



```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit

print("pandas: ", pd.__version__)
print("numpy: ", np.__version__)
print("matplotlib: ", matplotlib.__version__)
```

1. bargraph a.k.a (vertical bar)

The `plt.bar()` method allows us to plot the bar graph.

> You can get the bar clours from [colorhunt](https://colorhunt.co/).

```python
x = np.array(['Java', 'Kotlin', 'TypeScript', 'C++', 'Python'])
colors = np.array(['#86A789', '#B2C8BA', '#D2E3C8', '#EBF3E8', '#3A4D39'])
y = np.array([10, 20, 50, 3, 70])

plt.figure(figsize=(5, 3))
plt.bar(x, y, label=x, color=colors, width=.3)
plt.title('Programing Languages')
plt.xlabel("Programing Language", fontsize=8)
plt.ylabel("% likes", fontsize=8)
plt.legend()
plt.show()
```

Using the above data we want to plot a `horizontal` bar. To do that we will use a method calle `barh`

```python
plt.figure(figsize=(5, 3))
plt.barh(x, y, label=x, color=colors, height=.3)
plt.title('Programing Languages')
plt.ylabel("Programing Language", fontsize=8)
plt.xlabel("% likes", fontsize=8)
plt.legend()
plt.show()
```

2. linegraph

In the following code cell we are going to plot the line graph. To plot a line graph we use the `plot` method.

```python
x = y = np.arange(11)
plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=.8, linestyle='--', color='#0C356A', marker='o', markeredgecolor='red')
plt.plot(x**2, y, linewidth=.8, linestyle=':', color='#706233', marker='o', markeredgecolor='red')
plt.title('Line Graph', fontsize=12)
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.show()
```

3. scatterplot

In this section we are going to plot a scatter plot using  the `scatter` method.

```python
a = np.random.rand(10) * 100
b = np.random.rand(10) * 100


c = np.random.rand(10) * 100
d = np.random.rand(10) * 100

plt.figure(figsize=(5, 3))
plt.scatter(a, b,  color='#0C356A', marker='o', edgecolors='red')
plt.scatter(c, d, color='#4F6F52', marker='o', edgecolors='k')
plt.title('Scatter Plot', fontsize=12)
plt.xlabel("x", fontsize=8)
plt.ylabel("y", fontsize=8)
plt.show()
```

4. Pie chat

The method `pie` allows us to plot a `pie` chat. In the following code ceell we are going to plot the pie chat of  programing langaues with their likes.

```python
x = np.array(['Java', 'Kotlin', 'TypeScript', 'C++', 'Python'])
colors = np.array(['#86A789', '#B2C8BA', '#D2E3C8', '#EBF3E8', '#3A4D39'])
y = np.array([10, 20, 50, 3, 70])

plt.figure(figsize=(4, 4))
plt.pie(y, autopct="%1.2f%%", colors=colors, shadow=True, explode=[0.15, 0, 0, .2, 0]
       ,  textprops={'fontsize': 8, "color": 'red'}, radius=.8)
plt.legend(loc="upper right", fontsize=6, labels=x)
plt.title('Programing Languages', fontsize=10)
plt.show()
```

5. Box plot

Using the `.boxplot()` method we can plot a `boxplot`


```python
plt.figure(figsize=(5, 3))
plt.boxplot(y, flierprops=dict(markerfacecolor='r', marker='D'))
plt.title('Programing Languages', fontsize=10)
plt.show()
```

6. Histogram

The `hist` method can be used to plot a histogram.



```python
plt.figure(figsize=(4, 4))
plt.title('Programing Languages', fontsize=10)
plt.hist(y, cumulative=True, bins=5)
plt.show()
```

In the following code cell we are going to plot the line of best fit or a regression line from set of points `x` and `y`

```python
x = np.random.randint(5, 100, 10)
y = np.random.randint(0, 200, 10)
```

After that we are going to calculate the constant `c` and gradient `m` based on the equation of a straight line being given as:
    
``y = mx + c``

```python
c, m = polyfit(x, y, 1)
print(f"Gradient: {m}")
print(f"Y Intercept : {c}")
```

Let's create the `y` that we will plot againsit the `x` values for the line of best fit and store them in the `ys` variable.

```python
ys = m * x +  c
```

```python

plt.plot(x, ys, '-r', linewidth=1)
plt.rcParams["figure.figsize"] = (5, 3)
plt.scatter(x, y)
plt.title(
    "Scatter Plot", 
    fontfamily="Arial", 
    fontweight="bold", 
    fontsize=20
)
plt.xlabel(
    "x-axis",
    fontweight="bold",
    fontsize=15
)
plt.ylabel("y-axis", fontweight="bold", fontsize=15)
plt.show()
```

We can use the method called `savefig` to save a plot as an image.

```python

plt.plot(x, ys, '-r', linewidth=1)
plt.rcParams["figure.figsize"] = (5, 3)
plt.scatter(x, y)
plt.title(
    "Scatter Plot", 
    fontfamily="Arial", 
    fontweight="bold", 
    fontsize=20
)
plt.xlabel(
    "x-axis",
    fontweight="bold",
    fontsize=15
)
plt.ylabel("y-axis", fontweight="bold", fontsize=15)
plt.savefig('bestfit.png')
print('Saved')
plt.show()
```

**Questions:**

```python
df0 = pd.read_csv('data/airquality.csv')
df1 = pd.read_csv('data/HairEyeColor.csv')
```

```python
df0.head(1)
```

```python
df0.head(1)
```

We are going to drop null values.

```python
df0.dropna(inplace=True)
df1.dropna(inplace=True)
```

```python
# question a

wind = df0.Wind.values
temp = df0.Temp.values
c, m = polyfit(wind, temp, 1)
ys = m * wind + c

plt.figure(figsize=(5, 3))
plt.plot(wind, ys, '-k', linewidth=1)
plt.scatter(wind, temp,  color='#0C356A', marker='o', edgecolors='red')
plt.title('Wind aganist Temperature', fontsize=12)
plt.xlabel("wind", fontsize=8)
plt.ylabel("temp", fontsize=8)
plt.show()
```

```python
# question b
x = df0['Solar.R'].values[:10]
y = df0.Ozone.values[:10]

plt.figure(figsize=(5, 3))
plt.plot(x, y, linewidth=.8, linestyle='--', color='#0C356A', marker='o', markeredgecolor='red')
plt.title('Solar.R aganist Ozone', fontsize=12)
plt.xlabel("Solar.R ", fontsize=8)
plt.ylabel("Ozone", fontsize=8)
plt.show()
```

```python
# question c
freq = {v:0 for  v in df1.Hair.unique()}
for v in df1.Hair.values:
    freq[v] +=1
    
x = freq.keys()
colors = np.array(['#86A789', '#B2C8BA', '#D2E3C8', '#EBF3E8'])
y = freq.values()

plt.figure(figsize=(4, 4))
plt.pie(y, autopct="%1.2f%%", colors=colors, shadow=True
       ,  textprops={'fontsize': 8, "color": 'red'}, radius=.8)
plt.legend(loc="upper right", fontsize=6, labels=x)
plt.title('Hair Color', fontsize=10)
plt.show()
```

```python
# question d
freq = {v:0 for  v in df1.Sex.unique()}
for v in df1.Sex.values:
    freq[v] +=1
    
x = freq.keys()
colors = np.array(['#86A789', '#B2C8BA'])
y = freq.values()

plt.bar(x, y, label=x, color=colors, width=.3)
plt.title('Sex')
plt.xlabel("Gender", fontsize=8)
plt.ylabel("frequency", fontsize=8)
plt.legend()
plt.show()
```
