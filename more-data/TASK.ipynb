### Data Visualization

In this section we are going to visualize the data using the following visualization for all our dataframes

1. bar graphs
2. frequency table
3. pie chats
4. line graphs

Let's first define a function that will take in a `dictionary` of values paired with their respective counts and this function will draw a frequecy table from that dictionary.



````python
def get_freq_table(data: dict):
    table = PrettyTable()
    table.field_names = ["#", "Frequency"]
    table.add_rows([ d for d in data.items() ])
    print(table)
````

### Participants

In this section we are going to analyise the `participants` dataset. We are going how our dataset is looking like.

````python
participants_df.tail()
````

From the above we can create a count plot based on how many `Male` or `Female` `participants` attended. First we need to come up with the frequency or count of these numbers.

````python
gender_freq = {k: 0 for k in participants_df.sex.unique()}
for sex in participants_df.sex.values:
    gender_freq[sex] += 1
````

Now we can plot the frequecy table as follows.

````python
get_freq_table(gender_freq)
````

Base on the above frequecy table we can see that `Females` attended more than males. We can come up with a `piechat` from this.

````python
plt.figure(figsize=(10, 5))
plt.pie(gender_freq.values(), explode= [.0, .05], labels=gender_freq.keys(),
        pctdistance=0.6, shadow=True, autopct='%1.1f%%', textprops={'color': 'white', 'fontsize':10}, radius=.6)
plt.legend(fontsize=8)
plt.show()
````

### Clinic Codes

Let's try to visualize the clinic codes, that are comming from the `clinic_codes.csv` file. Just like what we did before we need to come up with the frequecy and the frequecy table.

````python
clinic_codes_df.head()
````

````python
clinic_codes_freq = {k: 0 for k in clinic_codes_df.clinic_code.unique()}
for code in clinic_codes_df.clinic_code.values:
    clinic_codes_freq[code] += 1
    
get_freq_table(clinic_codes_freq)
````

We can tell from the above frequecy table that most participants visites the clinic code `5` which has the highest frequecy of `741`. We can visualize this using a bar graph as follows:

````python
plt.figure(figsize=(5, 3), dpi=80)
plt.bar(clinic_codes_freq.keys(), clinic_codes_freq.values(), label= clinic_codes_freq.keys(), width=.5, align="center")
plt.title("A PLOT SHOWING HOW CLINIC CODES ARE DISTRIBUTED", color="k", fontsize=12)
plt.xlabel("Clinic Code", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
````

### Clinic Visits

In this dataset we are intrested in visualizing and see what was the patient's visit reason. Let's check the last `5` rows of our dataset.



````python
clinic_visits_df.tail()
````

Next we are going to create a frequecy table and display it which counts the purpose of visit of these patients to the hosipital.

````python
visitreason_freq = {k: 0 for k in clinic_visits_df.visitreason.unique()}
for reason in clinic_visits_df.visitreason.values:
    visitreason_freq[reason] += 1
get_freq_table(visitreason_freq)
````

We can see that most people where visiting the hosipital because of `Chronic Care - ART Follow-up` this is seen in the frequecy table because `Chronic Care - ART Follow-up` has high frequecy of `1, 512`. Nect we are going to visualize this using a bargraph. Inorder for us not to explode the plot we are going to create id's for the reasons of visit as intergers for example:

```
Chronic Care - Hypertension  -> id = 0
Chronic Care - Diabetes -> id = 1
...
```

````python
plt.figure(figsize=(8, 4), dpi=80)
ids = np.arange(len(visitreason_freq))
plt.bar(ids, visitreason_freq.values(), label= visitreason_freq.keys(), width=.5, align="center")
plt.title("A PLOT SHOWING REASONS FOR VISIT", color="k", fontsize=12)
plt.xticks(np.arange(min(ids), max(ids)+1, 1.0))
plt.xlabel("REASON ID", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
````
